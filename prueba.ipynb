$V_E = 0V$

$$V_{CE} = V_C - V_E = 12V$$

Ya que el transistor no está saturado:

$$I_C = 0A$$

Entonces el punto de operación de la \label{fig:1} es:

$$Q : (12V, 0A)$$

aplicando ley de tensiones de Kirchoff

$$R_{Th}I_B + V_{BE} + R_EI_E = V_{Th}$$

si $I_E = (\beta + 1)I_B$, entonces

$$R_{Th}I_B + V_{BE} + R_E(\beta + 1)I_B = V_{Th}$$

$$I_B = {V_{Th} - V{BE} \over R_{Th} + R_E(\beta + 1)}$$

Sustituyendo valores

$$I_B = {1.82V - 0.7V \over 8,48k\Omega + 200\Omega(100 + 1)} = 39.05\mu A$$

sabiendo que $I_C = \beta I_B$

$$I_C = 100(39.05\mu A) = 3.905mA$$

por otra parte

$$I_E = (\beta + 1)I_B = (100 + 1)39.05\mu A = 3.944mA$$

aplicando ley de tensiones de Kirchoff a la otra malla

$$R_{C}I_C + V_{CE} + R_EI_E = V_{CC}$$

$$V_{CE} = V_{CC} - R_{C}I_C - R_EI_E$$

sustituyendo

$$V_{CE} = 12V - 510\Omega (3.905mA) - 200\Omega (3.944mA) = 9.22V$$

por ley de ohm

$$V_E = R_EI_E = 200\Omega (3.944mA) = 0.79V$$

además

$$V_B = V_{BE} + V_E = 0.7V + 0.79V = 1.49V$$

$$V_C = V_{CE} + V_E = 9.22V + 0.79V = 10.01V$$

punto de operacion

$$Q = (9.22V, 3.905mA)$$
